```python
def format_tactics(tactics):
    formatted_tactics = []
    for tactic in tactics:
        words = tactic.split()
        if len(words) > 1:
            formatted_tactics.append('_'.join(words).upper())
        else:
            formatted_tactics.append(tactic.upper())
    return formatted_tactics
```

```python
df['Tactic'] = df['Tactic'].apply(lambda cell: ', '.join(format_tactics(eval(cell))))
```

```python
def extract_procedures_with_tactics(df, column_name):
    extracted_data = []
    
    for index, row in df.iterrows():
        if not pd.isna(row[column_name]):  # Check if the specified column is not NaN
            extracted_data.append((row[column_name], row['URL'] , row['Tactic']))
    
    extracted_df = pd.DataFrame(extracted_data, columns=[column_name, 'URL', 'Tactic'])
    
    return extracted_df
```

```python
result_df = extract_procedures_with_tactics(df, 'Description')
```

```python
import re
def remove_citations(text):
    pattern = r'\[\d+\]'

    preprocessed_text = re.sub(pattern, '', str(text))
    return preprocessed_text
#result_df['Description'] = result_df['Description'].apply(remove_citations)
result_df['Description'] = result_df['Description'].apply(remove_citations)
```
